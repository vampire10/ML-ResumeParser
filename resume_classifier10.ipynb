```python
#resume1_path = os.path.join(p, 'Resumes')
st = time.time()
resume_data_path = os.path.join(p, 'RESUME_DATA/BOT_RESUME_DUMP2000')
resume_folder_path = os.path.join(p, 'RESUME_DATA')
resume_list = os.listdir(resume_data_path)
print(len(resume_list))
java_points = []
cpp_points = []
kernel_points= []
python_points = []
data_str = []
#print(resume_list)
count = 0
for doc in resume_list:
    count += 1
    print(count)
    print(doc)
    path = os.path.join(resume_data_path, doc)
    data1 = textract.process(path).decode('utf-8').lower()
    data_str.append(data1)
    data = re.split("\W+", data1)
#    data = set(data1.split())
#    data = set(word_tokenize(data1))
    word_dict = Counter(data)
    #print(word_dict.most_common())
    j_word = set(word_dict.keys()).intersection(set(java))
    print(j_word)
    cp_word = set(word_dict.keys()).intersection(set(cpp))
    k_word = set(word_dict.keys()).intersection(set(kernel))
    py_word = set(word_dict.keys()).intersection(set(python))
    print((j_word, cp_word, k_word, py_word))    
#     java_value = len(java-data)
#     cpp_value = len(cpp-data)
#     kernel_value = len(kernel-data)
#     python_value = len(python-data)
    print([word for word in list(j_word) if word!=''])
    java_points.append(sum([word_dict[word] for word in list(j_word) if word!='']))
    cpp_points.append(sum([word_dict[word] for word in list(cp_word) if word!='']))
    kernel_points.append(sum([word_dict[word] for word in list(k_word) if word!='']))
    python_points.append(sum([word_dict[word] for word in list(py_word) if word!='']))
#     if doc=='vidya29.pdf':
#         break

    
ed = time.time()
print((ed-st)/60)
```
