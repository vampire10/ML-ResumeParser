```python
path_to_json = '/home/sohail/Desktop/DEskTop/resume_classification/RESUME_DATA/PARSED_RESUMES_ALL2000/'
resume_master_list = []
#sample_size = 395 #limit for testing

def find_email_from_json(resume):
    try:
        return ','.join(resume['basics']['email'])
    except:
        return ''
        
def find_education_from_json(resume):
    try:
        return str(resume['education_and_training'])
    except:
        return ''
    
def find_skills_from_json(resume):
    try:
        return str(resume['skills'])
    except:
        return ''
    
def find_work_and_academic_projects(resume):
    try:
        return str(resume['work_experience'])
    except:
        return ''
        
def find_education_with_percentage(resume):
    education_regex = r".*\d+[\.\d+]? *%.*"
    education_details = re.findall(education_regex, resume)
    return ','.join(education_details)
    
    
#main
sample_size = len(glob(path_to_json + '*.json'))
print(sample_size)
count = 0
for f_name in glob(path_to_json + '*.json'):
    print(f_name)
    if(sample_size == 0):
        break;
    
    count += 1
    print(count)
    fh = open(f_name, 'r')
    resume_data = fh.read()
 
    json_resume = json.loads(resume_data)
    
    #email
    email = find_email_from_json(json_resume)
    
    #education
    education_details = find_education_with_percentage(resume_data)
    if('' == education_details.strip()):
        education_details += find_education_from_json(json_resume)
        
    #skills
    skills = find_skills_from_json(json_resume)
    
    #work_and_prof_experience
    work_and_projects = find_work_and_academic_projects(json_resume)
    
    
    resume_extract = {'email': email, 
                      'education': education_details, 
                      'skills': skills, 
                      'file_name': f_name.split('/')[-1],
                      'work_and_projects': work_and_projects
                     }
    
    resume_master_list.append(resume_extract)
#dataframe
df = pd.DataFrame(resume_master_list)
df
```

```python
month_regex = re.compile(r"\b(?:Jan(?:uary)?|Feb(?:ruary)?|Mar(?:ch)?|Apr(?:il)?" + 
                         r"|May|Jun(?:e)?|Jul(?:y)?|Aug(?:ust)?|Sep(?:tember)?" + 
                         r"|Oct(?:ober)?|Nov(?:ember)?|Dec(?:ember)?)\b", re.IGNORECASE)
year_regex = re.compile(r"\b[1-3][0-9]{3}\b")
dict_keys_regex = re.compile(r"\"\b(.*?)\b\":")
special_chars_regex = re.compile(r"[^a-z. A-Z]")
excess_spaces_regex = re.compile(r"(?:\ ){2,}")
special_chars_regex_for_skills = re.compile(r"[^a-z. +#A-Z]")
http_link_regex = re.compile(r"(https?:\/\/(?:www\.|(?!www))[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}" +
                             r"|www\.[a-zA-Z0-9][a-zA-Z0-9-]+[a-zA-Z0-9]\.[^\s]{2,}|https?:\/\/(?:www\." + 
                             r"|(?!www))[a-zA-Z0-9]\.[^\s]{2,}|www\.[a-zA-Z0-9]\.[^\s]{2,})")
#dot_abbreviation_regex = re.compile(r"(\b(?:\w\.){3,})")
df['education'] = (df['education']
                   .replace({
                       percentage_regex: 'PERCENTAGE',
                       month_regex: 'MONTH',
                       year_regex: 'YEAR',
                       http_link_regex: 'HTTPLINK',
                       dict_keys_regex: ' ',
                       special_chars_regex: ' ',
                       excess_spaces_regex: ' '
                    })
                  )

df["skills"] = (df["skills"]
                .replace({
                    dict_keys_regex: ' ',
                    special_chars_regex_for_skills: ' ',
                    excess_spaces_regex: ' ' 
                })
                )

df["work_and_projects"] = (df['work_and_projects']
                           .replace({
                               http_link_regex: 'HTTPLINK',
                               month_regex: 'MONTH',
                               year_regex: 'YEAR',
                               dict_keys_regex: ' ',
                               special_chars_regex_for_skills: ' ',
                               excess_spaces_regex: ' '
                           })
                          )

#stemming
porter_stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))
stop_words.update(["percentage", "year", "month", "httplink"]) #add our custom added words
def stem_sentence(sentence):
    tokens = nltk.word_tokenize(sentence)
    result_tokens = [porter_stemmer.stem(token.lower()) for token in tokens if token.lower() not in stop_words]
    return ' '.join(result_tokens)

df['stemmed_education'] = df['education'].apply(stem_sentence)
df['stemmed_skills'] = df['skills'].apply(stem_sentence)
#df['stemmed_work_and_projects'] = df['work_and_projects'].apply(stem_sentence)
```
